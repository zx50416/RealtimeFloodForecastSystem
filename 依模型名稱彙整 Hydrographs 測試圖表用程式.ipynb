```python
import os
import shutil

def collect_test_graphs(model_name="XGBoost_AR"):
    source_root = f"./{model_name}"
    target_root = f"./{model_name}_graph"

    # å»ºç«‹è¼¸å‡ºä¸»è³‡æ–™å¤¾
    os.makedirs(target_root, exist_ok=True)

    for sub in sorted(os.listdir(source_root)):
        sub_path = os.path.join(source_root, sub)

        if not os.path.isdir(sub_path):
            continue

        hydro_path = os.path.join(sub_path, "Hydrographs")
        if not os.path.isdir(hydro_path):
            continue

        # å»ºç«‹å°æ‡‰è¼¸å‡ºå­è³‡æ–™å¤¾
        target_sub = os.path.join(target_root, sub)
        os.makedirs(target_sub, exist_ok=True)

        for file in os.listdir(hydro_path):
            if "test" in file.lower() and file.lower().endswith(".png"):
                src_file = os.path.join(hydro_path, file)
                dst_file = os.path.join(target_sub, file)

                if not os.path.exists(dst_file):
                    shutil.copy2(src_file, dst_file)

        print(f"âœ… å·²è™•ç†ï¼š{sub}")

    print(f"\nğŸ‰ åœ–è¡¨æ•´ç†å®Œæˆï¼š{target_root}")
```
