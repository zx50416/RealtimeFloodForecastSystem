```python
import os
import re

def rename_folders(prefix="tty", remove_number="1230"):
    current_dir = "."

    for name in os.listdir(current_dir):
        old_path = os.path.join(current_dir, name)

        # åªè™•ç†è³‡æ–™å¤¾
        if not os.path.isdir(old_path):
            continue

        # æ‰¾ç¬¬ä¸€å€‹æ•¸å­—çš„ä½ç½®
        match = re.search(r"\d", name)

        if match:
            first_digit_index = match.start()
            new_base_name = name[first_digit_index:]
        else:
            new_base_name = name

        # ğŸ”ª ç§»é™¤å¤šé¤˜çš„é€£çºŒæ•¸å­—ï¼ˆä¾‹å¦‚ 1230ï¼‰
        if remove_number in new_base_name:
            new_base_name = new_base_name.replace(remove_number, "")

        new_name = prefix + new_base_name
        new_path = os.path.join(current_dir, new_name)

        # é¿å…è¦†è“‹
        if new_name != name and not os.path.exists(new_path):
            os.rename(old_path, new_path)
            print(f"å·²é‡æ–°å‘½åï¼š{name} â†’ {new_name}")
        else:
            print(f"è·³éï¼š{name}")
```
