```python
import os
import re

# 請修改成你的 DATA 資料夾路徑
DATA_FOLDER = "./DATA/"

# 比對像是 IP10-1, IP28-3 這種
pattern = re.compile(r"^IP(\d+)-(\d+)$")

for folder in os.listdir(DATA_FOLDER):
    match = pattern.match(folder)
    if match:
        ip_num = int(match.group(1))  # 抓數字來判斷
        delta  = match.group(2)

        # 只修改 IP10 以上
        if ip_num >= 10:
            new_name = f"IP0{ip_num}-{delta}"
            old_path = os.path.join(DATA_FOLDER, folder)
            new_path = os.path.join(DATA_FOLDER, new_name)

            if old_path != new_path:
                print(f"重命名: {folder} -> {new_name}")
                os.rename(old_path, new_path)

print("✅ IP10 以後的資料夾名稱修改完成！")
```

```python
import os
import re

# 請修改成你的 DATA 資料夾路徑
DATA_FOLDER = "./DATA/"

# 比對像是 IP010-1, IP012-3 這種（前面多一個 0）
pattern = re.compile(r"^IP0(\d+)-(\d+)$")

for folder in os.listdir(DATA_FOLDER):
    match = pattern.match(folder)
    if match:
        ip_num = int(match.group(1))  # 抓數字
        delta  = match.group(2)

        # 只針對原本 >=10 的
        if ip_num >= 10:
            new_name = f"IP{ip_num}-{delta}"
            old_path = os.path.join(DATA_FOLDER, folder)
            new_path = os.path.join(DATA_FOLDER, new_name)

            if old_path != new_path:
                print(f"回復名稱: {folder} -> {new_name}")
                os.rename(old_path, new_path)

print("✅ IP010 以上的資料夾名稱已回復成 IP10 格式！")
```
